## DropNA (any)
Removing rows with __any__ NA's from the Dataset

### 53939 Rows
This shrinks the dataset a bit, but this may not be a great approach as we lose Vermont in the process, for some reason...

```python
dropna_any_geodist_long_CS= geodist_long_CS.dropna()

print(len(dropna_any_geodist_long_CS))
```

```python
dropna_any_geodist_long_CS.describe()
```

```python
print(len(dropna_any_geodist_long_CS["stateabb"].value_counts()))
print(dropna_any_geodist_long_CS["stateabb"].value_counts().sort_index())
```

## DropNA (ALL)
Removing rows where __all__ values are NA

### 825816 Rows

```python
dropna_all_geodist_long_CS= geodist_long_CS.dropna(how="all",subset=quant_cols)

print(len(dropna_all_geodist_long_CS))
```

```python
print(len(dropna_all_geodist_long_CS["stateabb"].value_counts()))
print(dropna_all_geodist_long_CS["stateabb"].value_counts())
```
